```python
#QUESTION 8

import pulp
import numpy as np
import matplotlib.pyplot as plt

# Define the linear programming problem
lp_problem = pulp.LpProblem("Maximize_Profit", pulp.LpMaximize)

# Define decision variables
x1 = pulp.LpVariable('x1', lowBound=0, cat='Continuous')
x2 = pulp.LpVariable('x2', lowBound=0, cat='Continuous')

# Objective function
lp_problem += 5*x1 + 3*x2, "Z"

# Constraints
lp_problem += 2*x1 + 3*x2 <= 60
lp_problem += 4*x1 + 2*x2 <= 80

# Solve the linear programming problem
lp_problem.solve()

# Extract optimal solution
optimal_x1 = x1.varValue
optimal_x2 = x2.varValue

# Print the optimal solution
print("Optimal solution:")
print("x1 =", optimal_x1)
print("x2 =", optimal_x2)
print("Maximum profit (Z) =", pulp.value(lp_problem.objective))

# Plot the feasible region
x1_values = np.linspace(0, 30, 400)
x2_values = np.linspace(0, 40, 400)
X1, X2 = np.meshgrid(x1_values, x2_values)
Z = 5*X1 + 3*X2

plt.figure(figsize=(8, 6))
plt.fill_between(x1_values, (60-2*x1_values)/3, color='blue', alpha=0.5)
plt.fill_between(x1_values, 0, (80-4*x1_values)/2, color='red', alpha=0.5)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Feasible Region')
plt.grid(True)
plt.xlim(0, 30)
plt.ylim(0, 40)
plt.show()
```
